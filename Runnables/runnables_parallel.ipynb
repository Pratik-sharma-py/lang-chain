```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
from langchain_cohere import ChatCohere
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableSequence,RunnableParallel
```

```python
load_dotenv()
```

```python
prompt1 = PromptTemplate(
    template="Write a tweet about {topic}",
    input_variables=['topic']
)
```

```python
prompt2 = PromptTemplate(
    template="Write a linkdin post about {topic}",
    input_variables=['topic']
)
```

```python
model = ChatCohere(model="command-r-plus-08-2024")
```

```python
parser = StrOutputParser()
```

```python
parallel_chain = RunnableParallel({
    'tweet': RunnableSequence( prompt1, model, parser ),
    'linkdin': RunnableSequence( prompt2, model, parser )
})
```

```python
result = parallel_chain.invoke({'topic':'Quantum Computing'})
```

```python
print(result['tweet'])
print(result['linkdin'])
```
