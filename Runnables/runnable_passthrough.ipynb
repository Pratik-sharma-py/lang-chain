```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
from langchain_cohere import ChatCohere
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableSequence,RunnableParallel,RunnablePassthrough
```

```python
load_dotenv()
```

```python
prompt1 = PromptTemplate(
    template="Write a joke about {topic}",
    input_variables=['topic']
)
```

```python
prompt2 = PromptTemplate(
    template="Explain the following joke {text}",
    input_variables=['text']
)
```

```python
model = ChatCohere(model="command-r-plus-08-2024")
```

```python
parsers = StrOutputParser()
```

```python
joke_gen_chain = RunnableSequence( prompt1, model, parsers )
```

```python
parallel_chain = RunnableParallel({
    'joke': RunnablePassthrough(),
    'explanation': RunnableSequence( prompt2, model, parsers)
})
```

```python
final_chain = RunnableSequence(joke_gen_chain, parallel_chain)
```

```python
result = final_chain.invoke({'topic':'WORLD'})
```

```python
print(result['joke'])
print(result['explanation'])
```
